#### EJERCICIO 2
Hacer un program en python que resuelva la ecuación cuadrática.

$\begin{equation}
p(x) = \frac {-b \pm \sqrt{b^2 - 4 \cdot a \cdot c}}{2 \cdot a}
\end{equation}$

````python
import math
# Entrada
a = int(input('Ingrese coeficiente a? '))
b = int(input('Ingrese coeficiente b? '))
c = int(input('Ingrese coeficiente c? '))
# Proceso
x1 = (-b + math.sqrt(b**2 - 4 * a * c))/(2*a)
x2 = (-b - math.sqrt(b**2 - 4 * a * c))/(2*a)
# Salida
print('x1: ',x1)
print('x2: ',x2)
````

#### EJERCICIO 3
Hacer una funcion en python que verificar la entrada de números enteros.

````python
import re
def numeroEntero(mensaje):
    patron = "[0-9]+"
    while True:
        n = input(mensaje)
        correcto = re.fullmatch(patron, n)
        if not correcto:
           print("Error vuelva a ingresar el dato")
        else:
           break
    return int(n)   
````

````python
x = numeroEntero('Ingresar número entero?')
y = 2 * x
print("Cuadrado: ", y)
````

````python
import numpy as np

# Crear un arreglo de NumPy
arreglo = np.array([1, 2, 3, 4, 5])

# Mostrar el arreglo
print("Arreglo original:", arreglo)

# Sumar un valor a cada elemento del arreglo
suma = arreglo + 10
print("Arreglo después de sumar 10:", suma)

# Calcular la suma total de los elementos del arreglo
suma_total = np.sum(arreglo)
print("Suma total de los elementos:", suma_total)

# Calcular la media de los elementos del arreglo
media = np.mean(arreglo)
print("Media de los elementos:", media)
````
