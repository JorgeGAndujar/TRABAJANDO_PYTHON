<span style="color:blue; font-size:15px">EJEMPLO 1</span><br>
Graficar la función trigonométrica seno:

$\begin{equation}
y = f(x) = \sin(x)
\end{equation}$

6000 MHz = 6000.000.000
1Hz = 1 ciclo/segundo

2 lecturas por cada  ciclo(subida(1) + bajada(1)) = 12.000.000.000 bits


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,360) #linspace te muestra la funcion que inicia en 0 hasta 360, y np.pi el numero pi
y = np.sin(x) # sin te provee la libreria de numpy
plt.plot(x,y,label="sin") # esta funcion .plot grafica
plt.title("FUNCION TRIGONOMÉTRICA SENO") # esta funcion .title pone el titulo
plt.xlabel("x")
plt.ylabel("y = sin(x)")
plt.legend()
plt.grid(True) # muestra una maya
plt.show()
```

<span style="color:blue; font-size:15px">EJEMPLO de frecuencia en la misma funcion</span><br>


```python
import numpy as np
import matplotlib.pyplot as plt
# metes la frecuencia(cantidad de ondas por segundo)
frecuencia = 3
x = np.linspace(0,2*np.pi,360) #linspace te muestra la funcion que inicia en 0 hasta 360, y np.pi el numero pi
y = np.sin(frecuencia * x) # sin te provee la libreria de numpy
plt.plot(x,y,label="sin") # esta funcion .plot grafica
plt.title("FUNCION TRIGONOMÉTRICA SENO") # esta funcion .title pone el titulo
plt.xlabel("x")
plt.ylabel("y = sin(x)")
plt.legend()
plt.grid(True) # muestra una maya
plt.show()
```

<span style="color:blue; font-size:15px">EJEMPLO 2</span><br>
Grafico de barras.

```python
import numpy as np
import matplotlib.pyplot as plt

meses = ["ENERO","FEBRERO","MARZO"] #eje x
ventas = [45384, 23435, 13434] # eje y

plt.figure(figsize=(12,6))
plt.bar(meses,ventas,color="lightgreen")
plt.title("VENTAS POR MES")
plt.xlabel("mes")
plt.ylabel("ventas")
plt.show()
```
